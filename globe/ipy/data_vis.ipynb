```python
import pandas as pd
import matplotlib.pyplot as plt

# Suppress scientific notation
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
# Open data, top 10 employer, and top 10 position csv files
employers = pd.read_csv('../data/top_employers.csv')
jobs = pd.read_csv('../data/top_jobs.csv')
data = pd.read_csv('../data/h1b_kaggle.csv')
```

```python
# Get wages filtering outliers
wages = data[data['PREVAILING_WAGE'] < 2000000000.0]['PREVAILING_WAGE'].dropna()
wages.describe()
```

```python
# Plot bar chart of wages
plt.plot(wages)
plt.savefig('../img/wage_breakdown.jpg')
```

```python
# Get breakdown of total applications by year
count = data[data['YEAR'] == 2011]['YEAR'].count()
year_counts = [count]
count = data[data['YEAR'] == 2012]['YEAR'].count()
year_counts.append(count)
count = data[data['YEAR'] == 2013]['YEAR'].count()
year_counts.append(count)
count = data[data['YEAR'] == 2014]['YEAR'].count()
year_counts.append(count)
count = data[data['YEAR'] == 2015]['YEAR'].count()
year_counts.append(count)
count = data[data['YEAR'] == 2016]['YEAR'].count()
year_counts.append(count)

# Plot breakdown by year
plt.scatter(['2011', '2012', '2013', '2014', '2015', '2016'], year_counts)
plt.savefig('../img/apps_by_year.jpg')
```

```python
# Get breakdown of certified applications by year
certified = data[data['CASE_STATUS'] == 'CERTIFIED']

count = certified[certified['YEAR'] == 2011]['YEAR'].count()
cert_counts = [count]
count = certified[certified['YEAR'] == 2012]['YEAR'].count()
cert_counts.append(count)
count = certified[certified['YEAR'] == 2013]['YEAR'].count()
cert_counts.append(count)
count = certified[certified['YEAR'] == 2014]['YEAR'].count()
cert_counts.append(count)
count = certified[certified['YEAR'] == 2015]['YEAR'].count()
cert_counts.append(count)
count = certified[certified['YEAR'] == 2016]['YEAR'].count()
cert_counts.append(count)

# Plot breakdown by year
plt.scatter(['2011', '2012', '2013', '2014', '2015', '2016'], cert_counts)
plt.savefig('../img/certs_by_year.jpg')
```
